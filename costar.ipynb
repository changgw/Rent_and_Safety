```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
crimes=pd.read_csv("data.csv")
```

```python
sns.countplot (x='BuildingRatingDesc',data=crimes,order=crimes['BuildingRatingDesc'].value_counts().index)
plt.xticks(rotation='vertical')
plt.ylabel('No of Building')
plt.savefig('C01.png')
```

```python
crimes.groupby(['BuildingRatingDesc','MinRent'])['NumBedsMin'].mean().unstack().plot(kind='bar')
plt.ylabel('min Price')
plt.show()
```

```python
sns.countplot (x='NumBedsMin',data=crimes,order=crimes['NumBedsMin'].value_counts().index)
plt.xticks(rotation='vertical')
plt.ylabel('NumBedsMin')
```

```python
crimes.groupby(['BuildingRatingDesc','MinRent'])
```

```python
crimes.groupby(['BuildingRatingDesc','MinRent'])['index']
```

```python
temp= crimes.pivot_table(values='index', index='BuildingRatingDesc', columns='NumBedsMax', aggfunc=np.size)
sns.heatmap(temp)
plt.ylabel('Building Rating')
plt.savefig('C03.png')
```

```python
temp= crimes.pivot_table(values='index', index='BuildingRatingDesc', columns='total', aggfunc=np.size)
sns.heatmap(temp)
plt.ylabel('Building Rating')
plt.xlabel('Total Services')
plt.savefig('C02.png')
```
